データ視覚化では、アンドリュース プロットまたはアンドリュース曲線は、高次元データの構造を視覚化する方法です。 
これは基本的に、ケント-キビアト レーダー m チャートのロールダウンされた非整数バージョン、または平行座標プロットの平滑化バージョンです。 
統計学者デビッド F. アンドリュースにちなんで命名されました。

また,2024年4月9日時点ではPolarsに対応していなかったので,元のコードのままです.

````python
import hvplot.pandas  # noqa
````

アンドリュース曲線は、多変量データのクラスターを視覚化するメカニズムを提供します。

アンドリュース曲線は関数形式を持っています:

```py
f(t) = x_1/sqrt(2) + x_2 sin(t) + x_3 cos(t) + x_4 sin(2t) + x_5 cos(2t) + ...
```

ここで、x 係数は各次元の値に対応し、t は -pi と +pi の間で直線的に配置されます。 フレームの各行は単一の曲線に対応します。

````python
from bokeh.sampledata import iris
iris = iris.flowers
````

````python
iris.head()
````

````python
hvplot.plotting.andrews_curves(
    iris,
    class_column = 'species',
    samples = 20,
)
````
