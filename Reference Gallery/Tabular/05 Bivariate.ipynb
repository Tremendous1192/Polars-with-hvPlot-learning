# Bivariate

````python
import hvplot.polars  # noqa
import polars as pl
````

```bivariate()``` は 2D 密度プロットを作成する統計手法です。
データが密すぎて各点を個別にプロットできない場合、二変量プロットは散布図の便利な代替手段となります。

````python
from bokeh.sampledata.autompg import autompg_clean as df

table = pl.from_pandas(df)
table.head()
````

````python
table.hvplot.bivariate('accel', 'mpg')
````
