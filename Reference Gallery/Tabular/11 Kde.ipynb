# Kde

````python
import hvplot.polars  # noqa
import polars as pl
````

Kernel density estimate (```kde```) provides a mechanism for showing the distribution and spread of the data.
In ```hvplot``` the method is exposed both as ```kde``` and ```density```.

````python
from bokeh.sampledata import sea_surface_temperature as sst

#df = sst.sea_surface_temperature
df = pl.from_pandas( sst.sea_surface_temperature.reset_index() )
df.tail()
````

````python
df["temperature"].hvplot.kde()
````

There are many options exposed and explorable using tab completion. In this case we’ll create a colormap that spans the rainbow and divide the temperature by month.

````python
import numpy as np
import colorcet as cc
````

````python
categorical_cmap = [cc.rainbow[int(i)] for i in np.linspace(0, 255, 12)]
````

````python
df = df.with_columns( pl.col("time").dt.month().alias("month") )
df.hvplot.kde("temperature", by = 'month', cmap = categorical_cmap, legend = 'top', height = 400)
````
