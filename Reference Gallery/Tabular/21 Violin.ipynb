# Violin

````python
import hvplot.polars  # noqa
import polars as pl
````

```violin``` plots are similar to box plots, but provide a better sense of the distribution of data.
Note that violin plots depend on the ```scipy``` library.

````python
from bokeh.sampledata.sprint import sprint

df = pl.from_pandas(sprint)
df.head()
````

````python
df.hvplot.violin(y='Time', by='Medal', c='Medal', ylabel='Sprint Time',
                 cmap=['gold', 'silver', 'brown'], legend=False, 
                 width=500, height=500, padding=0.4)
````
