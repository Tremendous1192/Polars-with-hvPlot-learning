````python
import hvplot.xarray  # noqa
````

**```image```** と同様に、```contour``` は 2D グリッド上に値を表示します。 ただし、まずデータをさまざまなレベルにセグメント化します。

````python
import xarray as xr

ds = xr.tutorial.open_dataset('air_temperature')
ds
````

````python
ds.hvplot.contour()
````

```contour``` プロットのスタイルと内容を制御するためのオプションが多数公開されています。

````python
(
    ds
    .mean(dim = 'time')# 全期間での気温の平均値で描画する
    .hvplot.contour(
        z = 'air', x = 'lon', y = 'lat', levels = 20,# 基本要素
        clabel = 'T [K]',# Z軸のラベル名
        label = 'Mean Air temperature [K]',
        cmap = 'gray'# 等高線のカラーマップ
    )
)
````

## Geographic Data
```tiles``` オプションを使用して、タイル サービスからのベースマップを含めます。

````python
(
    ds
    .hvplot.contour(
        geo = True,# 地図へのオーバーラップ
        tiles = 'EsriImagery',
        levels = 20,
        line_width = 2,
        cmap = 'reds',
    )
)
````
