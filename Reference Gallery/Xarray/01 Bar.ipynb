# 01 Bar

````python
import hvplot.polars  # noqa 
import xarray as xr
import polars as pl
````

## Introduction
```bar``` プロットは、高さが表す数値に比例した長方形の棒でカテゴリデータを表します。 X 軸はカテゴリを表し、Y 軸は数値スケールを表します。 バーは同じ幅なので、データを即座に比較できます。

## Data
Let’s load some data.

````python
ds = xr.tutorial.open_dataset('air_temperature').load()
air = ds.air
air1d = air.sel(lon=285.,lat=40.).groupby('time.month').mean()
air1d
````

````python
df = pl.from_pandas(air1d.to_dataframe())
df.head()
````

## Basic Bar Plots

````python
df.hvplot.bar(y = 'air', height = 500, title = "Air Temperature by Month")
````
